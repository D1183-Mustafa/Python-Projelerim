```python
from time import *
from random import *

print("""
10 sorudan oluşan 4 işlem oyununa Hoş geldiniz  :=)
1-100 sayı aralığında olan sorular sadece toplama, çıkarma , çarpma ve bölme işlemlerinden
oluşmaktadır.
        
Not : Hızlı Olmayı Unutma :) 
        
Bol Şans
        """)

sayac = 0
kontrol = 0
başlangıç = time()

while True:
    a = randint(1,100)
    b = randint(1,100)
    işlem = ["+", "-", "*", "/"]
    if sample(işlem,1)[0] == "+":
        sonuc = int(input(f"{a} + {b} :"))
        if sonuc == a + b:
            kontrol += 1
        sayac += 1
    elif sample(işlem,1)[0] == "-":
        sonuc = int(input(f"{a} - {b} :"))
        if sonuc == a - b:
            kontrol += 1
        sayac += 1
    elif sample(işlem,1)[0] == "*":
        sonuc = int(input(f"{a} * {b} :"))
        if sonuc == a + b:
            kontrol += 1
        sayac += 1
    elif sample(işlem,1)[0] == "/":
        b = randint(1,10)
        sonuc = float(input(f"{a} / {b} :"))
        if sonuc == a / b:
            kontrol += 1
        sayac += 1
    if sayac == 10:
        break
    
bitiş = time()

print("""

      Oyun bitti :) 
      Tüm soruları {:.2f} saniyede bitirdin
      Aldığın Puan {}""".format(bitiş-başlangıç, kontrol*10))
        
```
